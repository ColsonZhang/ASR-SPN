```python
from pathlib import Path
import librosa
import librosa.display
import IPython.display as ipd

import numpy as np
import matplotlib.pyplot as plt
```

```python
TRAIN_PATH = '../../dataset/trainset/'
labels = [ 'down', 'go', 'left', 'no', 'off', 'on', 'right', 'stop', 'up', 'yes']
```

```python
file_names = []

path = TRAIN_PATH + labels[0] + '/'
pathname = Path(path)

for cp in pathname.iterdir():
    term = Path(cp)
    file_names.append(term.name)
```

```python
def get_mfcc(test_file,hop_length=320,n_fft=2048,n_mfcc=13):
    # hop_length means the samples' number
    y, sr = librosa.load(test_file,sr=None)
    mfccs = librosa.feature.mfcc(y=y, sr=sr, hop_length=hop_length, n_mfcc=n_mfcc, n_fft=n_fft)
    return mfccs
```

```python
test_file = path + file_names[4]
mfccs = get_mfcc(test_file)
plt.figure(figsize=(15,5))
plt.imshow(mfccs, aspect='auto', origin='lower');
```

```python
test_file = path + file_names[1]
y, sr = librosa.load(test_file,sr=None)
plt.figure(figsize=(14, 5))
librosa.display.waveplot(y, sr=sr)
```
